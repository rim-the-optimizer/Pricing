$(1)$

$ \to d2_i$ is a new parameter: as $d_{ij}$ with $ (ij) \in E $ refers to all links possible, we can use only certain links for the objective function to calculate the income - $d2_i$ with $i \in A$: only the links starting in pick up tasks - the platform does not get any money for e.g. ODs driving from origin to destination
 
$ \to p_{ij}^h $ instead of $ p_i $: e.g. the price for a link from i=1 to j=2 and driver h=1 has possibly to be different than a price for a link i=1 to j=4 and driver number h=2 $ \to$ otherwise there will be no match allowed through constraint (16)    <br><br><br>

$(2) \to  j \in A \cup A' (if  \\\ \\\ \\\ i \neq j) $ instead of $ j \in A' $, because the links from a pickup to another pickup in a Multistop tour also has to be allowed and priced <br><br><br>

$(3) $

$ \to c_{ij}^h $ instead of $ c_{ij}$: costs for the platform depend also from the choice of the driver 

$ \to i \in \tau_h \cup A \cup A' $ and $ j \in A \cup A' (if  \\\ \\\ \\\ i \neq j) $ instead of $ i \in A $ and $ j \in A'$ : the driver is paid by the platform (costs) when he starts in his origin, drives to a pick up or drives to a delivery - but he will not be paid for his final way to his own destination - the former declaration does not include any costs for the way from the drivers' origins to the pick ups or any links between multiple pickups and deliveries <br><br><br>

$ (4) \to j \in A \cup A' $ instead of a missing domain of constraint $ (4) $: consideration of possible Multistops  <br><br><br>

$ (5) \to \sum_{h \in H}  \sum_{j \in A' \cup A (if \\\ j \neq i) } x^{h}_{ij} $ instead of $ \sum_{h \in H}  \sum_{j \in A'}x^{h}_{ij}$: consideration of links between two pickups <br><br><br>

$ (6) \to$ instead of using $V$ as the domain for the index $j$ the new Set $Vh_h$ is used, because we need $V$ in dependence of the driver $h$. Each driver has a different origin and a differnt destination, that is why $Vh_h$ is needed to distinguish between the referenced nodes for each driver. By this we can avoid x to be set to 1 in links which are not inteded for a driver ( a link to a destination of a different driver) <br><br><br>

$ (9) \to j \in A \cup A'$: consideration of multistop in flow conservation constraint (9) <br><br><br>

$ (12) \to i \in V $ is corrected to $ i \in A$: the domain for index $i$ was $V$ which cannot be feasible considering the right hand side of the constraint $S_{n+i}^h \\\ \to \\\ n+i$ may exceed the set of $V$ if $i \in V$ <br><br><br>

$ (16), (17), (18), (19) \to $ added domains $ \forall (i, j) \in E \\\ \\\ \\\ \forall h \in H$ as they were missing in the original model<br><br><br>

$ (25) $ is an additional constraint: $ \sum_{h \in H} \sum_{j \in V \ (if \ \tau_h \neq j)}x^{h}_{\tau_h,j} = 1   $ to make the routes start in the origins <br><br><br>
$ (26) $ is an additional constraint: $ \sum_{h \in H} \sum_{j \in V \ (if \ \tau_h' \neq i)}x^{h}_{\tau_h',j} = 0 $ to make the destinations the last possible nodes of each route <br><br><br>
